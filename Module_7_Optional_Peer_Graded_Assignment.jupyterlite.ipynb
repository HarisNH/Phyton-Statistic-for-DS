## Task 3 Use appropriate test to answer these questions

### 1. Is there a significant difference in median value of houses bounded by the Charles river or not? (T-test for independent samples)

1 if tract bounds river; 0 otherwise

*H0 is there is a significant difference in median value of houses bounded by the Charles river*

```python
## use MEDV and CHAS as variable
## testing if variance is equal of variance


# Levene's test for equal variances 
stat, p_value = scipy.stats.levene(boston_df[boston_df['CHAS'] == 1]['MEDV'], 
                                   boston_df[boston_df['CHAS'] == 0]['MEDV'], center='mean')

# Print the results
print(f"Levene's test statistic: {stat}")
print(f"P-value: {p_value}")

# Interpret the result
if p_value > 0.05:
    print("P-value > 0.05 thus we assume equal variances (fail to reject the null hypothesis).")
else:
    print("P-value < 0.05 thus we do not assume equal variances (reject the null hypothesis).")

```

```python
stat1, p_value1 = scipy.stats.ttest_ind(boston_df[boston_df['CHAS'] == 1]['MEDV'], 
                                        boston_df[boston_df['CHAS'] == 0]['MEDV'], equal_var = False)

# Print the results
print(f"T-test statistic: {stat1}")
print(f"P-value: {p_value1}")

# Interpret the result
if p_value1 > 0.05:
    print("P-value > 0.05 thus we assume equal variances (fail to reject the null hypothesis).")
else:
    print("P-value < 0.05 thus we do not assume equal variances (reject the null hypothesis).")
```

**Conclusion:** We reject H0 thus there might indicate no significant differences in median value of houses bounded by the Charles river because lack of evidence

## 2. Is there a difference in Median values of houses (MEDV) for each proportion of owner occupied units built prior to 1940 (AGE)? (ANOVA)

```python
# Using ANOVA
# Group by age
boston_df.loc[boston_df['AGE'] < 35, 'AGE_GROUP'] = 'lower_35'
boston_df.loc[(boston_df['AGE'] >= 35) & (boston_df['AGE'] <= 70), 'AGE_GROUP'] = 'between_35_70'
boston_df.loc[boston_df['AGE'] > 70, 'AGE_GROUP'] = 'older_70'

# Levene's test for equal variances
stat, p_value = scipy.stats.levene(boston_df[boston_df['AGE_GROUP'] == 'lower_35']['MEDV'], 
                                   boston_df[boston_df['AGE_GROUP'] == 'between_35_70']['MEDV'],
                                   boston_df[boston_df['AGE_GROUP'] == 'older_70']['MEDV'], 
                                   center='mean')

# Print the results
print(f"Levene's test statistic: {stat}")
print(f"P-value: {p_value}")

# Interpret the result
if p_value > 0.05:
    print("P-value > 0.05 thus we assume equal variances (fail to reject the null hypothesis).")
else:
    print("P-value < 0.05 thus we do not assume equal variances (reject the null hypothesis).")
```

```python
f_statistic, p_value = scipy.stats.f_oneway(lower_35, between_35_70, older_70)
print("F_Statistic: {0}, P-Value: {1}".format(f_statistic,p_value))

# Interpret the result
if p_value > 0.05:
    print("P-value > 0.05 thus we assume equal variances (fail to reject the null hypothesis).")
else:
    print("P-value < 0.05 thus we do not assume equal variances (reject the null hypothesis).")
```

**Conclusion:** Since the p-value is less than 0.05, we will reject the null hypothesis as there is significant evidence that at least one of the means differ.

## 3. No Relation between Nox and Non Retail Business per acre?

Can we conclude that there is no relationship between Nitric oxide concentrations and proportion of non-retail business acres per town? (Pearson Correlation)

```python
stat, p_value = scipy.stats.pearsonr(boston_df['NOX'], boston_df['INDUS'])

# Print the results
print(f"Levene's test statistic: {stat}")
print(f"P-value: {p_value}")

# Interpret the result
if p_value > 0.05:
    print("P-value > 0.05 thus we assume equal variances (fail to reject the null hypothesis).")
else:
    print("P-value < 0.05 thus we do not assume equal variances (reject the null hypothesis).")
```

**Conclusion:** Since the p-value  (Sig. (2-tailed)  < 0.05, we reject the Null hypothesis SO we conclude that there EXISTS a relationship between NOx gas concentration and Non Retail business per acre in town.

### 4. What is the impact of an additional weighted distance to the five Boston employment centres on the median value of owner occupied homes? (Regression analysis)

variabel DIS and MEDV

*   $H\_0: 1$ = 0 (additional weighted distance to the five Boston employment centres has NO EFFECT on the median value of owner occupied homes)
*   $H\_1: 1$ is not equal to 0 (additional weighted distance to the five Boston employment centres has an EFFECT on the median value of owner occupied homes)


```python
## X is the input variables (or independent variables)
X = boston_df['DIS']
## y is the target/dependent variable
y = boston_df['MEDV']
## add an intercept (beta_0) to our model
X = sm.add_constant(X) 

model = sm.OLS(y, X).fit()
predictions = model.predict(X)

# Print out the statistics
model.summary()
```

**Conclusion**: Like the t-test, the p-value is less than the alpha () level = 0.05, so we reject the null hypothesis as there is evidence that there is a an EFFECT on the median value of owner occupied homes
